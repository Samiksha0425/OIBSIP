```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
     
```

```python
sales_data = pd.read_csv('/content/Advertising.csv')
```

```python
sales_data.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
```

```python
sales_data.head(50)
```

```python
sales_data.info()
     
```

```python
sales_data.describe()
```

```python
sns.pairplot(sales_data)
```

```python
sns.heatmap(sales_data.corr(),annot = True)
```

```python
features = sales_data[["TV","Radio","Newspaper"]]
target = sales_data['Sales']
     
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.3,random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
linreg = LinearRegression() 
linreg.fit(x_train,y_train)
```

```python
coefficient = pd.DataFrame(linreg.coef_,features.columns,columns=['Coefficient'])
coefficient
```

```python
y_intercept = linreg.intercept_
y_intercept
     
```

```python
y_predict = linreg.predict(x_test)
y_predict
     
```

```python
plt.scatter(y_test,y_predict)
```

```python
sns.distplot((y_test,y_predict),bins=50)
```

```python
from sklearn import metrics
import math as math
print('MAE:',metrics.mean_absolute_error(y_test,y_predict))
print('MSE:',metrics.mean_squared_error(y_test,y_predict))
print('RMSE',math.sqrt(metrics.mean_squared_error(y_test,y_predict)))
     
```

```python
y_pred_test_df = pd.DataFrame({"Actual":y_test, "Predicted\n":y_predict})
y_pred_test_df
```
